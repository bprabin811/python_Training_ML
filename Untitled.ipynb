```python
import matplotlib.pyplot as plt
import numpy as np
N=int(input("Enter value of N:"))
array=[]
while N>1:
    if(N%2==0):
        N=N/2
        array.append(N)
    else:
        N=3*N+1
        array.append(N)
print(array)
p=np.array(array)
plt.plot(p, marker = 'o')
plt.show()
```
